```python
#Import the Drug_clean.csv file in DataResources Folder
#https://www.kaggle.com/datasets/thedevastator/drug-performance-evaluation?select=Drug_clean.csv
df = pd.read_csv('DataResources/Drug_clean.csv')

#Remove uncessary columns
df_clean = df.drop(columns=['Indication', 'Type'])

df_clean.head()
```

```python
# Conditions Table
# Determine the unique values in the Condition column
conditions = pd.DataFrame(df_clean['Condition'].unique(), columns=['Condition'])

# Include index values into the conditions df and rename column as Condition_no. This column will serve as PK.
conditions.reset_index(inplace=True)
conditions.rename(columns={'index': 'Condition_no'}, inplace=True)

#Save data into a CSV file
conditions.to_csv('DataResources/conditions.csv', index=False)
```

```python
#Drugs Table
# Determine the unique values in the Drug column
drugs = pd.DataFrame(df_clean['Drug'].unique(), columns=['Drug'])

# Include index values into the drugs df and rename column as Drugs_no. This column will serve as PK.
drugs.reset_index(inplace=True)
drugs.rename(columns={'index': 'Drugs_no'}, inplace=True)

#Save data into a CSV file
drugs.to_csv('DataResources/drugs.csv', index=False)
```

```python
#Forms Table

forms = pd.DataFrame(df_clean['Form'].unique(), columns=['Form'])

# Include index values into the forms df and rename column as Form_no. This column will serve as PK.
forms.reset_index(inplace=True)
forms.rename(columns={'index': 'Form_no'}, inplace=True)

#Save data into a CSV file
forms.to_csv('DataResources/forms.csv', index=False)
```

```python
import pandas as pd
from bokeh.plotting import figure, output_file, show
from bokeh.layouts import column
from bokeh.models import CustomJS, Slider, Select, HoverTool

# Convert DataFrame to ColumnDataSource
source = ColumnDataSource(data=filtered_data)

# Define interactive controls
conditions = Select(title="Condition", value="All", options=['All'] + sorted(filtered_data['Condition'].unique().tolist()))
price_slider = Slider(title="Max Price ($)", start=7, end=1000, value=250, step=1)
ease_slider = Slider(title="Minimum Ease of Use", start=1, end=5, value=5, step=1)

callback = CustomJS(args=dict(source=source, conditions=conditions, price_slider=price_slider, ease_slider=ease_slider), code="""
    const data = source.data;
    const condition = conditions.value;
    const price = price_slider.value;
    const ease = ease_slider.value;
    const drug = data['Drug'];
    const effective = data['Effective'];
    const price_data = data['Price'];
    const ease_data = data['EaseOfUse'];
    const condition_data = data['Condition'];
    const form_data = data['Drug Form'];
    const new_name = [];
    const new_effectiveness = [];
    const new_price = [];
    const new_ease = [];
    const new_condition = [];
    const new_form = [];
    for (var i = 0; i < drug.length; i++) {
        if ((condition === "All" || condition_data[i] === condition) &&
            price_data[i] <= price && ease_data[i] >= ease) {
            new_name.push(drug[i]);
            new_effectiveness.push(effective[i]);
            new_price.push(price_data[i]);
            new_ease.push(ease_data[i]);
            new_condition.push(condition_data[i]);
            new_form.push(form_data[i]);
        }
    }
    source.data['Drug'] = new_name;
    source.data['Effective'] = new_effectiveness;
    source.data['Price'] = new_price;
    source.data['EaseOfUse'] = new_ease;
    source.data['Condition'] = new_condition;
    source.data['Drug Form'] = new_form;
    source.change.emit();
""")


conditions.js_on_change('value', callback)
price_slider.js_on_change('value', callback)
ease_slider.js_on_change('value', callback)


# Create a plot
p = figure(title="Drug Effectiveness and Price", sizing_mode="stretch_width", height=400)
p.circle(x='Effective', y='Price', source=source, size=10, color="navy", alpha=0.5)
# Add x and y axes
p.xaxis.axis_label = "Effectiveness"  # Add label to x-axis
p.yaxis.axis_label = "Price"          # Add label to y-axis

# Create a HoverTool instance with the tooltip format
hover_tool = HoverTool(
    tooltips=[
        ("Drug Form", "@{Drug Form}"),
        ("Effectiveness", "@Effective"),
        ("Price", "@Price")
    ]
)

# Add the HoverTool to the plot
p.add_tools(hover_tool)

# Layout setup
layout = column(conditions, price_slider, ease_slider, p)

# Output HTML file
output_file("drug_dashboard.html")

# Show the plot
show(layout)
# print(source.data)
```
